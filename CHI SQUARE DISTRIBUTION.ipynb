```python
# CHI SQUARE DISTRIBUTION : IT IS BASICALLY USED AS A BASIS TO VERIFY THE HYPOTHESIS .\
# random.chisquare() IS USED FOR CHI SQUARE DISTRIBUTION .
# random.chisquare() HAS 2 ARGUMENTS : df() AND size() .
# df() MEANS DEGREE OF FREEDOM .

# GENERATE A CHI SQUARE DISTRIBUTION WITH DF OF 2 AND SIZE 4*6 .

from numpy import random
a = random.chisquare ( df=2 , size=(4,6) )
print(a)
```

```python
# GENERATE A CHI SQUARE DISTRIBUTION WITH DF OF 8 AND SIZE 4*6 .

from numpy import random
b = random.chisquare ( df=8 , size=(4,6) )
print(b)
```

```python
# VISUALIZATION OF CHI SQUARE DISTRIBUTION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

c = random.chisquare ( df=2 , size=400 )
sns.distplot ( c , hist=False )
plt.show()
```
