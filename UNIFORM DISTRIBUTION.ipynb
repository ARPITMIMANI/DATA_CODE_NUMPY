```python
# UNIFORM DISTRIBUTION : IT IS ONLY MADE FOR p() THAT IS PROBABILITY .
# random.uniform() IS USED FOR UNIFORM DISTRIBUTION .

# random.uniform() HAS 3 ARGUMENTS : a() , b() , size() .
# a() MEANS LOWER BOUND WHICH HAS A VALUE OF 0.0 .
# b() MEANS UPPER BOUND WHICH HAS A VALUE OF 1.0 .

# GENERATE UNIFORM DISTRIBUTION OF TWO DIMENSIONAL ARRAY .

from numpy import random
u = random.uniform ( size=(4,6))
print(u)
```

```python
# GENERATE UNIFORM DISTRIBUTION OF ONE DIMENSIONAL ARRAY .

from numpy import random
v = random.uniform ( size=(8))
print(v)
```

```python
# VISUALIZATION OF UNIFORM DISTRIBUTION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

w = random.uniform ( size=1000 )
sns.distplot ( w , hist=False )
plt.show()
```
