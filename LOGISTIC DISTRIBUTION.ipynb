```python
# LOGISTIC DISTRIBUTION : IT BASICALLY DESCRIBES GROWTH .
# LOGISTIC DISTRIBUTION IS VERY IMPORTANT IN REGRESSION AND NEURAL NETWORK .
# random.logistic() IS USED IN LOGISTIC DISTRIBUTION .

# random.logistic() HAS 3 ARGUMENTS : loc() , scale() , size() .
# loc() HAS A DEFAULT VALUE OF 0 .
# scale() HAS A DEFAULT VALUE OF 1 .

# GENERATE LOGISTIC DISTRIBUTION OF SIZE 4*6 WITH MEAN OF 8 AND STANDARD DEVIATION OF 10 .

from numpy import random
a = random.logistic ( loc=8 , scale=10 , size=(4,6) )
print(a)
```

```python
# GENERATE LOGISTIC DISTRIBUTION OF SIZE 4*4 

from numpy import random
b = random.logistic (  size=(4,4) )
print(b)
```

```python
# VISUALIZE LOGISTIC DISTRIBUTUION .

from numpy import random
import seaborn as sns
import matplotlib.pyplot as plt

c = random.logistic ( size=1000 )
sns.distplot ( c , hist=False )
plt.show()
```
