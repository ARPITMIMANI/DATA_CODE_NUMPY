```python
# RAYLEIGH DISTRIBUTION : IT IS USED IN SIGNAL PROCESSING .
# random.rayleigh() IS USED IN RAYLEIGH DISTRIBUTION .
# random.rayleigh() HAS 2 ARGUMENTS : scale() AND size() .
# scale() MEANS STANDARD DEVIATION , HOW FLAT IS THE DISTRIBUTION WITH DEFAULT VALUE OF 1.0 .

# GENERATE A RAYLEIGH DISTRIBUTION OF SCALE 2 AND SIZE 4*6 .

from numpy import random 
a = random.rayleigh ( scale=2 , size=(4,6) ) 
print(a)
```

```python
# GENERATE A RAYLEIGH DISTRIBUTION OF SCALE 8 AND SIZE 4*6 .

from numpy import random 
b = random.rayleigh ( scale=8 , size=(4,6) ) 
print(b)
```

```python
# VISUALIZATION OF RAYLEIGH DISTRIBUTION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

c = random.rayleigh ( scale=2 , size=800 ) 
sns.distplot ( c , hist=False )
plt.show()
```
