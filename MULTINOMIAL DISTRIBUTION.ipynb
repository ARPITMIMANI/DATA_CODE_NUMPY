```python
# MULTINOMIAL DISTRIBUION : IT A GENERALIZATION OF BINOMIAL DISTRIBUTION .
# random.multinomial() IS USED FOR MULTINOMIAL DISTRIBUION.

'''
random.multinomial() HAS 3 ARGUMENTS : n() , pvals() , size() .
n() MEANS NUMBER OF POSSIBILITY OR OUTCOME .
pvals() MEANS LIST OF POSSIBILITY OR OUTCOME .
size() MEANS SHAPE OF RETURNED ARRAY .
'''

# GENERATE A MULTINOMIAL DISTRIBUION FOR DICE ROLL OR CUBE SAMPLE .

from numpy import random
a = random.multinomial ( n=6 , pvals=[1/6,1/6,1/6,1/6,1/6,1/6] )
print(a)
```

```python
'''
MULTINOMIAL SAMPLES WILL NOT PRODUCE A SINGLE VALUE . 
THEY WILL PRODUCE ONE VALUE FOR EACH pvals() OR POSSIBILITY .
'''

# GENERATE A MULTINOMIAL DISTRIBUION FOR PYRAMIX SAMPLE .

from numpy import random
b = random.multinomial ( n=4 , pvals=[1/4,1/4,1/4,1/4] )
print(b)
```

```python
# GENERATE A MULTINOMIAL DISTRIBUION FOR COIN SAMPLE .

from numpy import random
c = random.multinomial ( n=2 , pvals=[1/2,1/2] )
print(c)
```

```python
# VISUALIZATION FOR MULTINOMIAL DISTRIBUION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

a = random.multinomial ( n=6 , pvals=[1/6,1/6,1/6,1/6,1/6,1/6] )
sns.distplot ( a , hist=False )
plt.show()
```
