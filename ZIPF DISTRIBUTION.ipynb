```python
# ZIPF DISTRIBUTION : IT IS THE PROBABILTY OF COMMON OBJECTS .
# random.zipf() IS USED FOR ZIPF DISTRIBUTION .
# random.zipf() HAS 2 ARGUMENTS : a() AND size() .

# GENRATE A ZIPF DISTRIBUTION WITH SHAPE 2 AND SIZE 4*6 .

from numpy import random 
e = random.zipf ( a=2 , size=(4,6) )
print(e)
```

```python
# VISUALIZATION OF ZIPF DISTRIBUTION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

k = random.zipf ( a=2 , size=800 )
sns.distplot ( k , kde=False )
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

k = random.zipf ( a=2 , size=800 )
sns.distplot ( k[k<=10] , kde=False )
plt.show()
```
