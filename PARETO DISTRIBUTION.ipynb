```python
# PARETO DISTRIBUTION : IT IS ALSO KNOWN AS 80:20 RATIO DISTRIBUTION .
# 80:20 RATIO DISTRIBUTION MEANS 20% FACTORS CAUSE 80% OUTCOME OR RESULT .
# random.pareto() IS USED FOR PARETO DISTRIBUTION .
# random.pareto() HAS 2 ARGUMENTS : a() AND size() .
# a() MEANS SHAPE OF DISTRIBUTION .

# GENERATE A PARETO DISTRIBUTION OF SHAPE 2 AND SIZE 4*6 .

from numpy import random
e = random.pareto ( a=2 , size=(4,6) )
print(e)
```

```python
# GENERATE A PARETO DISTRIBUTION OF SHAPE 8 AND SIZE 4*6 .

from numpy import random
f = random.pareto ( a=8 , size=(4,6) )
print(f)
```

```python
# VISUALIZATION OF PARETO DISTRIBUTION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

g = random.pareto ( a=2 , size=800 )
sns.distplot ( g , kde=False )
plt.show()
```
