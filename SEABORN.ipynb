```python
# seaborn() : IT IS A LIBRARY THAT USES matplotlib() UNDERNEATH TO PLOT GRAPH THAT IS pyplot() .
# distplot() : IT STANDS FOR DISTRIBUTION PLOT .
# DISTRIBUTION PLOT IS CURVE PLOT LIKE HISTOGRAM .

# PLOTTING A distplot() WITH HISTOGRAM .

import matplotlib.pyplot as plt
import seaborn as sns

a = [0,1,2,3,4,5]
sns.distplot(a)
plt.show()
```

```python
# PLOTTING A distplot() WITHOUT HISTOGRAM .

import matplotlib.pyplot as plt
import seaborn as sns

a = [0,1,2,3,4,5]
sns.distplot ( a , hist=False )
plt.show()
```
