```python
# EXPONENTIAL DISTRIBUTION : IT IS USED FOR DESCRIBING THE TIME TILL NEXT EVENT THAT IS FAILURE OR SUCCESS .
# random.exponential() IS USED FOR EXPONENTIAL DISTRIBUTION .
# random.exponential() HAS TO ARGUMENTS scale() AND size() .
# scale() IS INVESRSE OF RATE OR lam() .

# GENERATE A EXPONENTIAL DISTRIBUTION OF SCALE OF 2 AND SIZE OF 4*6 .

from numpy import random
a = random.exponential ( scale=2  , size=(4,6) )
print(a)
```

```python
# GENERATE A EXPONENTIAL DISTRIBUTION OF SCALE OF 8 AND SIZE OF 4*6 .

from numpy import random
b = random.exponential ( scale=8  , size=(4,6) )
print(b)
```

```python
# VISUALIZATION OF EXPONENTIAL DISTRIBUTION .

from numpy import random 
import matplotlib.pyplot as plt
import seaborn as sns

c = random.exponential ( size=400 ) 
sns.distplot ( c , hist=False )
plt.show()
```
