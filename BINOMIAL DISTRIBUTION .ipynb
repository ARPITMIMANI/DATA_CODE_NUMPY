```python
# BINOMIAL DISTRIBUTION IS A DISCRETE DISTRIBUTION WITH BINARY OUTPUT .
# random.binomial() IS USED FOR BINOMIAL DISTRIBUTION .

'''
random.binomial() HAS 3 PARAMETERS : n() , p() , size() .
n() IS NUMBER OF TRIALS .
p() IS PROBABILITY .
size() IS SHAPE OF RETURNED ARRAY.
'''

# GIVEN 10 TRIAL FOR A COIN WILL GENERATE 10 DATA POINTS .

from numpy import random
a = random.binomial ( n=10 , p=0.5 , size=10 )
print(a)
```

```python
# VISUALIZATION OF BINOMIAL DISTRIBUTION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

f = random.binomial ( n=10 , p=0.5 , size=(1000) )
sns.distplot ( f , hist=True  , kde=False )
plt.show()
```

```python
''' 
THE DIFFERENECE BETWEEN NORMAL DISTRIBUTION AND BINOMIAL DISTRIBUTION IS THAT
- NORMAL DISTRIBUTION IS A CONTINUOUS DISTRIBUTION  
- BINOMIAL DISTRIBUTION IS A DISCRETE DISTRIBUTION .
'''

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

g = random.normal ( loc=40 , scale=20 , size=1000 )
h = random.binomial ( n=200 , p=0.2 , size=1000 )

sns.distplot ( g , hist=False , label="NORMAL" )
sns.distplot ( h , hist=False , label="BINOMIAL" )

plt.show()
```
