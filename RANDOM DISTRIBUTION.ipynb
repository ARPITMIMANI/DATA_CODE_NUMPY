```python
# NORMAL DISTRIBUTION IS ALSO KNOWN AS GAUSSIAN DISTRIBUTION .
# random.normal() IS USED FOR NORMAL DISTRIBUTION .
# random.normal() HAS 3 ARGUMENTS : loc() , scale() , size() .
# loc() IS KNOWN AS MEAN .
# scale() IS KNOWN AS STANDARD DEVIATION .
# GENERATE A RANDOM NORMAL DISTRIBUTION OF SIZE 4*6 .

from numpy import random
a = random.normal ( size=(4,6) )
print(a)
```

```python
# GENERATE A RANDOM NORMAL DISTRIBUTION OF SIZE 4*6  WITH MEAN OF 3 AND STANDARD DEVIATION OF 2 .

from numpy import random 
b = random.normal ( loc=3 , scale=2 , size=(4,6) )
print(b)
```

```python
# THE CURVE OF NORMAL DISTRIBUTION IS ALSO KNOWN AS BELL CURVE .
# VISUALIZATION OF NORMAL DISTRIBUTION WITOUT OUT HISTOGRAM .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot ( random.normal(size=(1000)) , hist=False )
plt.show()
```
