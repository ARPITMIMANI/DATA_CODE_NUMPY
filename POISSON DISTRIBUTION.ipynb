```python
# POISSON DISTRIBUTION : IT ESTIMATES HOW MANY TIMES EVENT CAN HAPPEN .
# POISSON DISTRIBUTION IS A DISCRETE DISTRIBUTION .

# random.poisson() IS USED FOR POISSON DISTRIBUTION .
# random.poisson() HAS 2 ARGUENTS : lam() , size() .
# lam() MEANS NUMBER OF OCCURANCE OR RATE .

# GENERATE RANDOM POISSON DISTRIBUTION OF SIZE 10 AND NUMBER OF OCCURANCE 2 .

from numpy import random
a = random.poisson ( lam=2 , size=10 )
print(a)
```

```python
# GENERATE RANDOM POISSON DISTRIBUTION OF SIZE 10 AND NUMBER OF OCCURANCE 2 .

from numpy import random
b = random.poisson ( lam=2 , size=10 )
print(b)
```

```python
# VISUALIZATION OF POISSON DISTRIBUTION .

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

c = random.poisson ( lam=2 , size=1000 )
sns.distplot ( c , kde=False )
plt.show()
```

```python
''' 
THE DIFFERENECE BETWEEN NORMAL DISTRIBUTION AND POISSON DISTRIBUTION IS THAT
- NORMAL DISTRIBUTION IS A CONTINUOUS DISTRIBUTION  
- POISSON DISTRIBUTION IS A DISCRETE DISTRIBUTION .
'''

# NOTE : n() * p() == lam()

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

d = random.normal ( loc=40 , scale=8 , size=1000 )
e = random.poisson ( lam=40 , size=1000 )

sns.distplot ( d , hist=False , label="NORMAL" )
sns.distplot ( e , hist=False , label="POISSON" )

plt.show()
```
